```python
import pandas as pd

# Ensure full text in DataFrame is shown for screenshot
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.width', 1000)

intents = [
    "check order status",
    "reset password",
    "technical issue",
    "refund request",
    "connect to human agent"
]

responses = {
    "check order status": "Please provide your order number so I can check the status for you.",
    "reset password": "You can reset your password here: https://example.com/reset",
    "technical issue": "Can you describe the technical issue you're facing in more detail?",
    "refund request": "Please submit your refund request using this form: https://example./refund",
    "connect to human agent": "Sure, connecting you to a human agent now..."
}

df = pd.DataFrame({
    "Intent": intents,
    "Response": [responses[intent] for intent in intents]
})

print(df.head())
```

```python
from transformers import pipeline

# Load zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Raw Input (Before)
user_input = "I want to know the status of my order"
intents = ["check order status", "reset password", "technical issue", "refund request", "connect to human agent"]


result = classifier(user_input, intents)

print("\n=== AFTER ===")
print("Top Predicted Intent:", result["labels"][0])
print("Scores:", list(zip(result["labels"], result["scores"])))

```

```python
import matplotlib.pyplot as plt

intent_counts = {intent: 1 for intent in intents}  # Each defined once

plt.bar(intent_counts.keys(), intent_counts.values(), color='skyblue')
plt.title("Intent Distribution (Manual)")
plt.xlabel("Intent")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
from transformers import pipeline
import matplotlib.pyplot as plt

# Zero-shot pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Example user input
query = "I forgot my password and need help logging in"
intents = ["check order status", "reset password", "technical issue", "refund request", "connect to human agent"]

# Run the model
result = classifier(query, intents)

# Extract labels and scores
labels = result["labels"]
scores = result["scores"]

# Plot confidence scores
plt.figure(figsize=(8, 4))
plt.barh(labels[::-1], scores[::-1], color='green')
plt.title("Model Confidence Scores for Query")
plt.xlabel("Confidence Score")
plt.tight_layout()
plt.show()
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

intents = ["check order status", "reset password", "technical issue", "refund request", "connect to human agent"]

test_query = "I need to know where my package is"
result = classifier(test_query, intents)

print("Input:", test_query)
print("Top Intent:", result["labels"][0])
print("Confidence Scores:")
for label, score in zip(result["labels"], result["scores"]):
    print(f"{label}: {score:.4f}")
```
